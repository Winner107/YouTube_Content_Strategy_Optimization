```python
import pandas as pd
import numpy as np
```

```python
comments_df = pd.read_csv('Documents/comments.csv')
video_df = pd.read_csv('Documents/videos-stats.csv')
```

```python
video_df.head()
```

```python
comments_df
```

```python
video_df.columns = video_df.columns.str.lower().str.replace(' ', '_')
comments_df.columns = comments_df.columns.str.lower().str.replace(' ', '_')
```

```python
video_df
```

```python
perfomance_df = video_df.merge(comments_df, how='left', on='video_id')
perfomance_df
```

```python
perfomance_df.isnull().sum()
```

```python
perfomance_df = perfomance_df.drop(['unnamed:_0_x', 'unnamed:_0_y'], axis=1)
perfomance_df
```

## Performance Percentile

```python
perfomance_df['published_at'] = pd.to_datetime(perfomance_df['published_at'])
perfomance_df
```

```python
perfomance_df = perfomance_df.sort_values('published_at')
```

```python
perfomance_df['views_percentile_7d'] = perfomance_df['views']\
    .rolling(7, min_periods=1).apply(lambda s: s.rank(pct=True).iloc[-1])
```

```python
perfomance_df
```

```python
perfomance_df['views_percentile_30d'] = perfomance_df['views']\
    .rolling(30, min_periods=1).apply(lambda s: s.rank(pct=True).iloc[-1])
```

```python
perfomance_df['views_percentile_all'] = perfomance_df['views'].rank(pct=True)
```

```python
perfomance_df = perfomance_df.rename(columns={'comment':'comment_text'})
perfomance_df
```

## Sentiment Analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
comments_df['sentiment_score'] = comments_df['comment'].apply(lambda x: sia.polarity_scores(str(x))['compound'])
comments_df
```

```python
video_sentiment = comments_df.groupby('video_id')['sentiment_score'].mean().reset_index()
video_sentiment
```

```python
perfomance_df = perfomance_df.merge(video_sentiment, on='video_id', how='left')
perfomance_df
```

```python
perfomance_df = perfomance_df.rename(columns={'likes_x':'comment_likes', 'likes_y':'video_likes'})
perfomance_df
```

## Engagement Score

```python
perfomance_df['engagement_score_in_million'] = (perfomance_df['views'] + perfomance_df['video_likes'] + perfomance_df['comments']) / 1000000
```

```python
perfomance_df
```

## Sentiment-Performance Correlation

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=perfomance_df, x='sentiment_score', y='engagement_score_in_million')
plt.title('Sentiment vs Engagement Score')
plt.xlabel('Average Comment Sentiment')
plt.ylabel('Engagement Score')
plt.show()

corr = perfomance_df[['sentiment_score', 'engagement_score_in_million']].corr().iloc[0,1]
print(f"Correlation between sentiment and engagement: {corr:.2f}")
plt.savefig('Sentiment_vs_Engagement_Score.png')
```

## Publish Month Cohort Analysis

```python
perfomance_df['published_month'] = perfomance_df['published_at'].dt.to_period('M')
perfomance_df.head()
```

```python
perfomance_df['cumulative_views_in_million'] = (perfomance_df.groupby('video_id')['views'].cumsum()) / 1000000
perfomance_df.head()
```

```python
perfomance_df['days_since_publish'] = (perfomance_df['published_at'].max() - perfomance_df['published_at']).dt.days
perfomance_df.head()
```

```python
cohort_views = perfomance_df.groupby(['published_month', 'days_since_publish'])['cumulative_views_in_million'].mean().reset_index()
cohort_views.head()
```

```python
plt.figure(figsize=(10,6))
for month in cohort_views['published_month'].unique():
    subset = cohort_views[cohort_views['published_month'] == month]
    plt.plot(subset['days_since_publish'], subset['cumulative_views_in_million'], label=str(month))

plt.title('Average Cumulative Views by Publish Month')
plt.xlabel('Days Since Publish')
plt.ylabel('Average Cumulative Views')
plt.show()
```

## Engagement Quality Monitor

```python
perfomance_df['flag_for_review'] = ((perfomance_df['sentiment_score'] < -0.2) & 
                                    (perfomance_df['engagement_score_in_million'] > perfomance_df['engagement_score_in_million'].median()))
```

```python
perfomance_df.head()
```

```python
flagged_videos = perfomance_df[perfomance_df['flag_for_review']].drop_duplicates(subset='video_id')
print(f"Total videos flagged for review: {len(flagged_videos)}")
```
